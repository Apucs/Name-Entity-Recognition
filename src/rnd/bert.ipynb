```python
import pandas as pd
import numpy as np
data = pd.read_csv("conllpp_up_train.csv",encoding="latin1" )

data.head(10)
```

```python
data.insert(0, 'sentence', np.nan)
```

```python
data.head()
```

```python
indexes = data.index[data['word']=='.'].tolist()

# indexes_to_in = []

# for ind in indexes:
#   indexes_to_in.append(ind-1)

# indexes_to_in
```

```python
val = 0
for id in indexes:
  data.loc[data.index[id], 'sentence'] = val
  val = val+1
```

```python
data.head(50)
```

```python
data =data.fillna(method ="ffill")
data.head(50)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data["sentence"] = LabelEncoder().fit_transform(data["sentence"] )
data.tail(50)
```

```python
data.rename(columns={"sentence":"sentence_id","word":"words","O_BILOU":"labels"}, inplace =True)
```

```python
data["labels"] = data["labels"].str.upper()
```

```python
X= data[["sentence_id","words"]]
Y =data["labels"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size =0.2)
```

```python
x_train
```

```python
train_data = pd.DataFrame({"sentence_id":x_train["sentence_id"],"words":x_train["words"],"labels":y_train})
test_data = pd.DataFrame({"sentence_id":x_test["sentence_id"],"words":x_test["words"],"labels":y_test})
```

```python
train_data
```

```python
from simpletransformers.ner import NERModel,NERArgs
```

```python
label = data["labels"].unique().tolist()
label
```

```python
args = NERArgs()
args.num_train_epochs = 3
args.learning_rate = 1e-4
args.overwrite_output_dir =True
args.train_batch_size = 32
args.eval_batch_size = 32
```

```python
model = NERModel('bert', 'bert-base-cased',labels=label,args =args)
```

```python
model.train_model(train_data,eval_data = test_data,acc=accuracy_score)
```

```python
from datasets import load_dataset

dataset = load_dataset("conllpp")
```

```python
result, model_outputs, preds_list = model.eval_model(test_data)
```
